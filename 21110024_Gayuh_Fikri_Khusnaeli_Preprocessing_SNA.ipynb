```python
#import modul
import pandas as pd
import numpy as np
import re
import re as reg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Membaca data dari file Excel
data=pd.read_excel("ReviewTerbaru.xlsx", sheet_name='Lokawisata Baturraden')
data
```

# PREPROCESSING

```python
!pip install Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Preprocessing
slangs = {'yg': 'yang', 'tdk': 'tidak', 'pd': 'pada', 'mlh': 'malah',
          'jgn': 'jangan', 'jg': 'juga', 'tp': 'tapi', 'blkg': 'belakang',
          'dr': 'dari', 'klo': 'kalo', 'lg': 'lagi'}
preprocessing_review = []

for sentence in data['Review']:
    # Hapus semua karakter khusus
    processed_comment = re.sub(r'\W', ' ', str(sentence))

    # Konversi menjadi huruf kecil
    processed_comment = processed_comment.lower()

    # Hapus angka
    processed_comment = re.sub(r'\d+', ' ', processed_comment)

    # Hapus kata-kata yang terdiri dari satu, dua, atau tiga huruf
    processed_comment = re.sub(r'\b\w{1,3}\b', '', processed_comment)

    # Hapus karakter duplikat
    pattern = re.compile(r"(.)\1{1,}", re.DOTALL)
    processed_comment = pattern.sub(r"\1", processed_comment)

    # Koreksi kata-kata slang
    words = processed_comment.split()
    rfrm = [slangs[word] if word in slangs else word for word in words]
    processed_comment = " ".join(rfrm)

    # Hapus stopword
    factory = StopWordRemoverFactory()
    more_stopword = ['tak', 'jd', 'per', 'nya']  # Menambahkan stopword
    stopwords = factory.get_stop_words() + more_stopword
    temp = [t for t in re.findall(r'\b[a-z]+-?[a-z]+\b', processed_comment) if t not in stopwords]
    processed_comment = ' '.join(temp)

    # Stemming
    stemmer = StemmerFactory().create_stemmer()
    processed_comment = stemmer.stem(processed_comment)

    # Ganti beberapa spasi dengan satu spasi
    processed_comment = re.sub(r'\s+', ' ', processed_comment, flags=re.I)

    # Tambahkan komentar yang memiliki lebih dari 3 huruf ke dalam list
    if len(processed_comment.split()) > 3:
        preprocessing_review.append(processed_comment)

# Buat DataFrame untuk data yang telah diproses
preprocessing_review = pd.DataFrame({'Processed_Review': preprocessing_review})
```

```python
# Tampilkan DataFrame hasil preprocessing
print(preprocessing_review)
```

# MENYIMPAN KE EXCEL

```python
# Simpan data yang telah diproses ke dalam file Excel
preprocessing_review.to_excel("Processed_Review.xlsx", index=False)
```

```python
# Jumlah record/dokumen sebelum dan sesudah pembersihan data
jumlah_record_sebelum = len(data)
jumlah_record_sesudah = len(preprocessing_review)
print("Jumlah record/dokumen sebelum pembersihan data:", jumlah_record_sebelum)
print("Jumlah record/dokumen setelah pembersihan data:", jumlah_record_sesudah)
```
